```python
from openai import OpenAI
import time

client = OpenAI(
  base_url="https://openrouter.ai/api/v1",
  api_key="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",
)

max_retries = 3
retry_delay_seconds = 5

for attempt in range(max_retries):
    try:
        completion = client.chat.completions.create(
          extra_headers={
            "HTTP-Referer": "<YOUR_SITE_URL>", # Optional. Site URL for rankings on openrouter.ai.
            "X-Title": "<YOUR_SITE_NAME>", # Optional. Site title for rankings on openrouter.ai.
          },
          extra_body={},
          model="mistralai/mistral-7b-instruct:free", # Changed model here
          messages=[
            {
              "role": "user",
              "content": "how non technical learner transit towards datasciece?"
            }
          ]
        )
        print(completion.choices[0].message.content)
        break # If successful, break the loop
    except Exception as e:
        # Check if the error is a RateLimitError and if retries are available
        if "RateLimitError" in str(e) and attempt < max_retries - 1:
            print(f"Rate limit hit. Retrying in {retry_delay_seconds} seconds... (Attempt {attempt + 1}/{max_retries})")
            time.sleep(retry_delay_seconds)
        else:
            print(f"An unexpected error occurred or maximum retries reached: {e}")
            raise # Re-raise if not a rate limit error or max retries exceeded
```
